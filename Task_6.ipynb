### **Problem Statement 1: Smart Parking Lot Management System**

```python
def smart_parking_system(capacity , logs):
  current_vehicles =0
  peak_usage =0

  #Parked vehicle calculation logic
  for log in logs:
    if log == "IN":
      current_vehicles +=1
    elif log == "OUT":
      if current_vehicles > 0:
        current_vehicles -= 1

    #peak_usage = The maximum number of vehicles that were parked at any time during the day
    if current_vehicles > peak_usage:
       peak_usage = current_vehicles

  print(f"Currently Parked Vehicles: {current_vehicles}")
  print(f"Peak parking usage : {peak_usage}")

  #Capacity logic
  if current_vehicles > capacity :
    print("Parking Status: Exceeded Capacity! Alert!")
  else:
    print("Parking Status: Available")

capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking_system(capacity, vehicle_logs)
```

### **Problem Statement 2: Online Food Delivery Time Estimator**

```python
def delivery_time_estimator(distance, traffic, weather):
    base_time = 20

    travel_time = (distance / 25) * 60   # in minutes    # Convert distance to delivery time (speed = 25 km/hr)

    # Traffic conditions
    if traffic == "Low":
        traffic_delay = 0
    elif traffic == "Medium":
        traffic_delay = 10
    elif traffic == "High":
        traffic_delay = 20
    else:
        traffic_delay = 0

    # Weather conditions
    if weather == "Clear":
        weather_delay = 0
    elif weather == "Cloudy":
        weather_delay = 5
    elif weather == "Rainy":
        weather_delay = 10
    elif weather == "Storm":
        weather_delay = 20
    else:
        weather_delay = 0

    total_time = base_time + travel_time + traffic_delay + weather_delay      #calculation => 20 + 19.2 + 20 + 10 = 69.2 minutes

    print("Estimated Delivery Time:",round(total_time), "minutes")


# Sample Input
delivery_time_estimator(8, "High", "Rainy")
```

### **Problem Statement 3:Movie Theatre Seat Occupancy Analyzer**

```python
def seat_occupancy_analyzer(total_seats, booked_seats):

    # Count booked seats
    booked_count = len(booked_seats)

    # Calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:",round(occupancy),"%")

    # Conditions
    if occupancy == 100:
        print("Show Status: Housefull")
        print("Suggestion: Open additional show")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")


# Sample Input
total_seats = 200
booked_seats = [1] * 150   # 150 booked seats

seat_occupancy_analyzer(total_seats, booked_seats)
```

### **Problem Statement 4:Cloud Server Load Classification System**

```python
def classify_server_load(cpu_readings):

    # Calculate total using loop
    total = 0
    for reading in cpu_readings:
        total += reading

    # Calculate average
    average = total / len(cpu_readings)

    print("Average CPU Load:", round(average),"%")

    # Classification
    if average < 50:
        print("Server Status: Normal")
    elif 50 <= average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
cpu_data = [45, 60, 70, 85, 90]
classify_server_load(cpu_data)
```

### **Problem Statement 5:Smart Classroom Resource Usage Monitor**

```python
def monitor_resources(resource_usage):
    overused = []

    # Checking each resource usage
    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)

    # Displaying results
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resources(usage)
```

### **Problem Statement 6: Online Event Registration Capacity Controller**

```python
def manage_registrations(capacity, total_registrations):

    confirmed = 0
    waitlisted = 0

    for i in range(1, total_registrations + 1):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    if confirmed >= capacity:
        status = "Closed"
    else:
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)

capacity = 100
registrations = 105

manage_registrations(capacity, registrations)
```
