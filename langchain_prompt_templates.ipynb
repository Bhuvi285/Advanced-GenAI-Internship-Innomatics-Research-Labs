## ✅ Task 1: Replace Hardcoded Prompts (10%)

**Goal:** Convert a plain f-string function into a reusable LangChain `PromptTemplate`.

````python
# ❌ Old approach – hardcoded f-string (DO NOT USE)
# def explain_topic(topic):
#     return f"Explain {topic} in simple terms for beginners"

# ✅ New approach – LangChain PromptTemplate
explain_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

# Function to generate prompt using the template
def explain_topic(topic):
    """
    Generates a beginner-friendly explanation prompt using PromptTemplate.
    Args:
        topic (str): The topic to explain
    Returns:
        str: The formatted prompt string
    """
    return explain_template.format(topic=topic)

# Test
print("┦ Task 1 Output:")
print(explain_topic("Machine Learning"))
print(explain_topic("Neural Networks"))
print(explain_topic("Python"))
````

## ✅ Task 2: Multi-Input Prompt System (15%)

**Goal:** Build a template that takes multiple inputs – `topic`, `audience`, and `tone`.

````python
# Multi-input PromptTemplate
multi_input_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

# Function to generate multi-input prompt
def multi_input_prompt(topic, audience, tone):
    """
    Generates a prompt tailored to the audience and tone.
    Args:
        topic (str): Subject to explain
        audience (str): Target audience
        tone (str): Tone of explanation
    Returns:
        str: Formatted prompt
    """
    return multi_input_template.format(topic=topic, audience=audience, tone=tone)

# Test cases
test_cases = [
    ("AI", "beginners", "friendly"),
    ("Python", "kids", "fun"),
    ("Deep Learning", "engineers", "technical")
]

print("🔹 Task 2 Output:")
for topic, audience, tone in test_cases:
    print(f"→ {multi_input_prompt(topic, audience, tone)}")
````

## ✅ Task 3: Prompt Variations Engine (15%)

**Goal:** Create 3 different PromptTemplates – Teaching, Interview, Storytelling – and run all for the same topic.

````python
# Define 3 PromptTemplate variations
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 questions about {topic}"
)

storytelling_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

# Store all templates in a dictionary for clean access
prompt_variations = {
    "Teaching": teaching_template,
    "Interview": interview_template,
    "Storytelling": storytelling_template
}

# Function to run all variations for a given topic
def run_prompt_variations(topic):
    """
    Runs all 3 prompt templates for the given topic.
    Args:
        topic (str): Topic to generate prompts for
    """
    print(f"📚 Prompt Variations for topic: '{topic}'\n")
    for style, template in prompt_variations.items():
        prompt = template.format(topic=topic)
        print(f"  [{style}] → {prompt}")

# Run for Machine Learning
print("🔹 Task 3 Output:")
run_prompt_variations("Machine Learning")
````

## ✅ Task 4: ChatPromptTemplate System (15%)

**Goal:** Build a chat-based prompt system using System + User roles, supporting multiple role behaviors.

````python
# Define role-based system messages
role_system_messages = {
    "teacher": "You are a patient and clear teacher who explains concepts step by step.",
    "interviewer": "You are a technical interviewer who asks insightful questions to evaluate understanding.",
    "motivator": "You are an enthusiastic motivator who explains topics with energy and encouragement."
}

# Function to build a ChatPromptTemplate dynamically based on role
def build_chat_prompt(role, topic):
    """
    Builds a ChatPromptTemplate for the given role and topic.
    Args:
        role (str): One of 'teacher', 'interviewer', 'motivator'
        topic (str): The topic to discuss
    Returns:
        list: List of formatted chat messages
    """
    # Validate role
    if role not in role_system_messages:
        raise ValueError(f"Invalid role '{role}'. Choose from: {list(role_system_messages.keys())}")

    # Build the ChatPromptTemplate
    chat_template = ChatPromptTemplate.from_messages([
        ("system", role_system_messages[role]),
        ("human", "Please help me understand: {topic}")
    ])

    # Format and return the messages
    messages = chat_template.format_messages(topic=topic)
    return messages

# Test with teacher + Neural Networks
print("🔹 Task 4 Output:")
roles_to_test = ["teacher", "interviewer", "motivator"]
topic = "Neural Networks"

for role in roles_to_test:
    messages = build_chat_prompt(role, topic)
    print(f"\n🎭 Role: {role.upper()}")
    for msg in messages:
        print(f"  [{msg.__class__.__name__}]: {msg.content}")
````

## ✅ Task 5: Input Validation Layer (10%)

**Goal:** Create a `validate_inputs()` function that checks audience and tone against allowed values.

````python
# Define allowed values
VALID_AUDIENCES = ["beginner", "intermediate", "expert"]
VALID_TONES = ["formal", "casual", "fun"]

# Default fallback values
DEFAULT_AUDIENCE = "beginner"
DEFAULT_TONE = "casual"

def validate_inputs(audience, tone):
    """
    Validates audience and tone inputs.
    - If valid, returns them as-is.
    - If invalid, assigns default values with a warning.

    Args:
        audience (str): Target audience
        tone (str): Tone of the explanation
    Returns:
        tuple: (validated_audience, validated_tone)
    """
    validated_audience = audience
    validated_tone = tone

    # Validate audience
    if audience.lower() not in VALID_AUDIENCES:
        print(f"⚠️  Invalid audience '{audience}'. Defaulting to '{DEFAULT_AUDIENCE}'.")
        validated_audience = DEFAULT_AUDIENCE

    # Validate tone
    if tone.lower() not in VALID_TONES:
        print(f"⚠️  Invalid tone '{tone}'. Defaulting to '{DEFAULT_TONE}'.")
        validated_tone = DEFAULT_TONE

    return validated_audience.lower(), validated_tone.lower()

# Test cases
print("🔹 Task 5 Output:")

# Case 1: All valid
print("\n[Test 1] Valid inputs:")
print(validate_inputs("beginner", "fun"))

# Case 2: Invalid audience
print("\n[Test 2] Invalid audience:")
print(validate_inputs("student", "casual"))

# Case 3: Invalid tone
print("\n[Test 3] Invalid tone:")
print(validate_inputs("expert", "sarcastic"))

# Case 4: Both invalid
print("\n[Test 4] Both invalid:")
print(validate_inputs("anyone", "random"))
````

## ✅ Task 6: Prompt Generator App (15%)

**Goal:** Create a `generate_prompt()` function that combines validation, style selection, and prompt generation.

````python
# Style-based PromptTemplates
style_templates = {
    "teaching": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} in a {tone} teaching style, step by step"
    ),
    "interview": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Ask 3 insightful interview questions about {topic} suitable for {audience} in a {tone} tone"
    ),
    "storytelling": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} as a {tone} story that a {audience} can understand and enjoy"
    )
}

# Valid styles list
VALID_STYLES = list(style_templates.keys())

def generate_prompt(topic, audience, tone, style):
    """
    Full prompt generator with validation and style selection.
    Args:
        topic (str): Subject to explain
        audience (str): Target audience (beginner/intermediate/expert)
        tone (str): Tone of explanation (formal/casual/fun)
        style (str): Prompt style (teaching/interview/storytelling)
    Returns:
        str: Final formatted prompt
    """
    # Step 1: Validate audience and tone
    audience, tone = validate_inputs(audience, tone)

    # Step 2: Validate style
    if style.lower() not in VALID_STYLES:
        print(f"⚠️  Invalid style '{style}'. Defaulting to 'teaching'.")
        style = "teaching"

    # Step 3: Select the appropriate template
    selected_template = style_templates[style.lower()]

    # Step 4: Generate and return the prompt
    return selected_template.format(topic=topic, audience=audience, tone=tone)

# Test outputs
print("🔹 Task 6 Output:")

# Example from assignment
print("\n[Example 1]:")
print(generate_prompt("Neural Networks", "beginners", "fun", "storytelling"))

print("\n[Example 2]:")
print(generate_prompt("Python", "intermediate", "casual", "teaching"))

print("\n[Example 3]:")
print(generate_prompt("Deep Learning", "expert", "formal", "interview"))

# Example with invalid inputs to show validation in action
print("\n[Example 4 - with invalid inputs]:")
print(generate_prompt("AI Ethics", "students", "weird", "teaching"))
````

## ✅ Task 7: Template Reusability Test (10%)

**Goal:** Use ONE PromptTemplate and run it with 5 different inputs to prove reusability.

````python
# Single reusable PromptTemplate
reusable_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

# 5 different inputs to test reusability
reusability_test_cases = [
    {"topic": "Machine Learning",  "audience": "beginners",    "tone": "friendly"},
    {"topic": "Cloud Computing",   "audience": "professionals", "tone": "formal"},
    {"topic": "Blockchain",        "audience": "kids",          "tone": "fun"},
    {"topic": "Data Structures",   "audience": "students",      "tone": "casual"},
    {"topic": "Generative AI",     "audience": "executives",    "tone": "concise"}
]

print("🔹 Task 7 Output – Reusability Test (Same Template, 5 Different Inputs):")
print("=" * 65)

for i, inputs in enumerate(reusability_test_cases, start=1):
    # Use the SAME template each time
    prompt = reusable_template.format(**inputs)
    print(f"\n[Run {i}] → {prompt}")

print("\n✅ Same template structure used across all 5 runs – Reusability confirmed!")
````

## 🔁 Full Pipeline Summary

```
User Input → Validation → Prompt Template Selection → Dynamic Prompt Generation → Output
```

````python
# Full end-to-end pipeline demo
def full_pipeline(topic, audience, tone, style):
    """
    Demonstrates the complete prompt engineering pipeline:
    User Input → Validation → Template Selection → Prompt Generation → Output

    Args:
        topic (str): Subject to explain
        audience (str): Target audience
        tone (str): Tone of explanation
        style (str): Prompt style
    """
    print("📥 Step 1: User Input Received")
    print(f"   Topic: {topic} | Audience: {audience} | Tone: {tone} | Style: {style}")

    print("\n🔍 Step 2: Validating Inputs...")
    validated_audience, validated_tone = validate_inputs(audience, tone)
    print(f"   Validated → Audience: {validated_audience} | Tone: {validated_tone}")

    print("\n📋 Step 3: Selecting Prompt Template...")
    selected_style = style if style in style_templates else "teaching"
    print(f"   Style selected: {selected_style}")

    print("\n⚙️  Step 4: Generating Dynamic Prompt...")
    final_prompt = style_templates[selected_style].format(
        topic=topic,
        audience=validated_audience,
        tone=validated_tone
    )

    print("\n✅ Step 5: Final Output")
    print(f"   {final_prompt}")
    return final_prompt

# Run the full pipeline
print("🔹 Full Pipeline Demo:")
print("=" * 55)
full_pipeline("Transformers", "beginner", "fun", "storytelling")
````
