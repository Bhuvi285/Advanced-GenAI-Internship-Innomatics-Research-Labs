### **Problem 1: Social Media – Post Engagement Analyzer**

```python
def analyze_engagement(likes_list):
  total_likes = 0

  for likes in likes_list:
    total_likes += likes

  if(total_likes > 1000):
    status = "Viral Post"
  else:
    status = "Normal Engagement"

  print(f"Total Likes:{total_likes}")
  print(f"Post Status:{status}")

likes_data = [200 , 150 , 300 , 400]
analyze_engagement(likes_data)
```

### **Problem 3: Agriculture – Rainfall Adequacy Checker**

```python
def rainfall_checker(rainfall_data , required_level):
  total_rainfall = 0

  for rain in rainfall_data:
     total_rainfall += rain               #Calculate total rainfall first and then go for average rainfall

  avg_rainfall = total_rainfall/ len(rainfall_data)

  if avg_rainfall >= required_level:
    status = "Adequate Rainfall"
  else:
    status = "Inadequate Rainfall"

  print(f"Average Rainfall:{int(avg_rainfall)}")
  print(f"Rainfall Status: {status}")

season_data = [ 70 , 75 , 80 , 63]
required = 70

rainfall_checker(season_data , required)
```

### **Problem 4: Social Media – Duplicate Account Detection**

```python
def duplicate_detection(usernames):
  unique_usernames = set()
  duplicate_found = False

  for user in usernames:
    if user in unique_usernames:
      duplicate_found = True
      break
    else:
      unique_usernames.add(user)

  if duplicate_found:
    print(f"Duplicate Accounts Found: Yes")
  else:
    print("Duplicate Accounts Found: No")

users = ["rahul123", "neha_01", "amit99", "rahul123"]
duplicate_detection(users)
```

### **Problem 9: E-Learning – Student Attendance Eligibility System**

```python
def attendence_eligibility(att_list):
  total_classes = len(att_list)

  present_count = 0

  for status in att_list:
    if status == "P":
      present_count += 1
    else:
      continue

  attendence_percentage = (present_count / total_classes) * 100

  if attendence_percentage >= 75:
    status = "Eligible"
  else:
    status = "Not Eligible"

  print(f"Attendance Percentage: {attendence_percentage}")
  print(f"Exam Eligibility: {status}")

att_list = ["P" , "P" , "P" , "A" , "P" ]
attendence_eligibility(att_list)
```

### **Problem 10: Smart Electricity Bill Analyzer**

```python
def calculate_bill(units):   #calculating total bill for units first

  if units <= 100:
    return units * 3
  elif units <= 200:
    return (100 * 3) + ((units - 100) * 5)
  else:
    return (100 * 3) +  (( 100 * 5)) + ((units - 200) * 7)

def usage_calculator(total_bill):  #Estimating total usage

  if total_bill < 500 :
    return "Low Usage"
  elif total_bill <= 1500:
    return "Moderate Usage"
  else:
    return "High Usage"

total_bill = calculate_bill(250)
print(f"Total Bill = {total_bill}")
Usage_Analysis = usage_calculator(total_bill)
print(f"Usage Analysis = {Usage_Analysis}")
```
