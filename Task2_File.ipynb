### **Problem 1: Unique Words in a Sentence**

```python
sentence = input("Enter a sentence: ")

words = sentence.split()    #Split the sentence into words

unique_words = set(words)     #Convert the list into set to remove duplicates

print("Total count of unique words:" , len(unique_words))
print("Unique words:" , unique_words)
```

### **Problem 2: Highest Salary from Employee Data**

```python
employees ={
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

max_salary = 0
max_employee = ""

for name , salary in employees.items():   # .items() returns pairs (tuples) , Python automatically unpacks the tuple name = 'Ravi' , salary = 75000.
                                          # This is called tuple unpacking.
  if salary > max_salary:
    max_salary = salary
    max_employee = name

print(f"Highest salary: {max_employee} - {max_salary}")
```

### **Problem 3: Find Maximum and Minimum Values**

```python
numbers = [45, 22, 89, 10, 66]

max_num = numbers[0]
min_num = numbers[0]

i = 1

for i in numbers:
  if i > max_num:
    max_num = i
  if i < min_num:
    min_num = i

print(f"Max: {max_num}")
print(f"Min: {min_num}")
```

### **Problem 4: Count Products Above a Price Threshold**

```python
prices = [450, 1200, 899, 1500, 300]

count = 0

for i in prices:
  if i > 1000:
    count = count + 1
print(f"Products above 1000: {count}")
```

### **Problem 5: Calculate Attendance Percentage**

```python
attendance = ["P", "P", "A", "P", "P"]

count = 0
for i in attendance:
  if i == "P":
    count += 1
avg_percentage = (count / len(attendance)) * 100

print(f"Attendance Percentage: {avg_percentage}%")
```

### **Problem 7: Count Character Frequency**

```python
text = "pypthon"

char_count = {}

for i in text:
  if i in char_count:
    char_count[i] += 1
  else:
    char_count[i] = 1

print(char_count)
```

### **Problem 9: Check if a Key Exists in a Dictionary**

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

key = input("Enter a key:")

if key in employees:
  print("Employee exists")
else:
  print("Employee does not exist")
```

### **Problem 10: Calculate Average Marks**

```python
marks = [65, 70, 80, 75]

sum = 0
for i in marks:
  sum += i
avg = sum / len(marks)

print(f"Average Marks: {avg}")
```
