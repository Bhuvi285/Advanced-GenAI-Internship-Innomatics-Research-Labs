### **Problem 1: Mobile Recharge Validation System**

```python
#Create a function to validate a mobile recharge request.

def mobile_recharge(amount):
  valid_plans = [199, 299, 399, 599]

  if amount < 50:
    print("The recharge amount must be greater than or eqal to 50rs")
    return False
  elif amount not in valid_plans:
    print("Invalid plans Selected. Available Plans : 199rs, 299rs, 399rs, 599rs")
  else:
    print("Recharge Suscessful")
    return True

while True:

  #try : try this code if it fails let the except handle it correctly
  try:
       amount = int(input("Enter recharge amount :"))

       if mobile_recharge(amount):
        break
       else:
        print("Please try again.\n")

  except ValueError:
    print("Please enter a valid number.\n")
```

### **Problem 2: Inventory Reorder Alert System**

```python
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(f"{product} -> Reorder Alert (Stock: {stock})")
        else:
            print(f"{product} -> Stock OK (Stock: {stock})")


# Example inventory dictionary
inventory = {
    "Rice": 25,
    "Sugar": 10,
    "Oil": 8,
    "Soap": 30,
    "Salt": 14
}

check_inventory(inventory)
```

### **Problem 3: Student Result Processing System**

```python
def process_result(marks):
    total = 0

    # Loop to calculate total
    for mark in marks:
        total += mark

    average = total / len(marks)

    print(f"Average Marks: {average}")

    if average >= 50:
        print("Result: Pass")
        return "Pass"
    else:
        print("Result: Fail")
        return "Fail"


# Example usage
student_marks = [60, 45, 70, 55]
result = process_result(student_marks)

print("Returned Value:", result)
```

### **Problem 4: Cab Fare Estimator with Retry Option**

```python
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12

    fare = base_fare + (distance * per_km_rate)

    if peak.lower() == "yes":
        fare = fare + (fare * 0.25)   # 25% extra

    return fare


# Retry System
while True:
    try:
        distance = float(input("Enter distance in km: "))

        if distance <= 0:
            print(" Distance must be greater than 0.\n")
            continue

        peak = input("Is it peak hour? (yes/no): ")

        total_fare = calculate_fare(distance, peak)

        print(f"Total Fare: ₹{total_fare}")
        break   # Exit loop if successful

    except ValueError:
        print("Please enter valid numeric distance.\n")
```

### **Problem 5: Employee Attendance Eligibility Checker**

```python
def check_attendance(attendance_list):
    present_days = 0

    # Loop to count present days
    for day in attendance_list:
        if day == "P":
            present_days += 1

    total_days = len(attendance_list)

    if total_days == 0:
        return "No Attendance Data"

    percentage = (present_days / total_days) * 100

    print(f"Attendance Percentage: {percentage}%")

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Example usage
attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]
result = check_attendance(attendance)

print("Eligibility Status:", result)
```

### **Problem 6: Password Strength Checker**

```python
def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    # Check minimum length
    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"

    # Loop over string
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    # Final condition check
    if has_digit and has_special:
        return "Strong Password "
    else:
        return "Weak Password (Must include digit and special character)"


# Example usage
pwd = input("Enter password: ")
result = check_password(pwd)
print(result)
```
