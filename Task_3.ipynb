### **Problem Statement 1: Employee Performance Bonus Eligibility**

```python
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}

highest = max(employees.values()) #Extracts all scores.

top_performers = []

for name , score in employees.items():
  if score == highest :
    top_performers.append(name)

print("Top Performers Eligible for Bonus:" , ",".join(top_performers), f"(Score: {highest})")
# ", ".join(top_performers) => Converts list into a comma-separated string.
```

### **Problem Statement 2: Search Query Keyword Analysis**

```python
query = "Buy mobile phone buy phone online"

query = query.lower()


#importing string for removing punctuations
import string
for char in string.punctuation:
  query = query.replace(char , "")

words = query.split()      # splits the query into the list

word_count = {}

for word in words:
  if word in word_count:
    word_count[word] += 1
  else:
      word_count[word] = 1

result = {}

for word , count in word_count.items():
  if count > 1:
    result[word] = count
print(result)
```

### **Problem Statement 3: Sensor Data Validation**

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid = []

for index, value in enumerate(sensor_readings):       #enumerate automatically returns index , value instead of applying the traditional method
  if value % 2 == 0:
    valid.append((index , value))

print("Valid Sensor Readings (Hour, Value):")
print(valid)
```

### **Problem Statement 4: Email Domain Usage Analysis**

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

for email in emails:
  domain = email.split("@")[1]                       #Ex- split : ["ravi", "gmail.com"]
  domain_count[domain] = domain_count.get(domain , 0) + 1  #Give me the current value of this domain.
                                                              #If it doesn’t exist, assume it is 0.
                                                                #Then add 1 and store it back.

total = len(emails)

for domain , count in domain_count.items():
  print(f"{domain}: {int((count/total)*100)}%")

```

### **Problem Statement 5: Sales Spike Detection**

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

avg = sum(sales) / len(sales)

threshold = avg * 1.3 #testing value for 30% more than average sales

for index , value in enumerate(sales):
  if value > threshold:
    print(f"Day {index + 1}: {value}")
```

### **Problem Statement 6: Duplicate User ID Detection**

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

user_count = {}

for user in user_ids:
  user_count[user] = user_count.get(user , 0) + 1  #Give me current count of this user. If this user doesn’t exist, assume count is 0.

for user , count in user_count.items():
  if count > 1:
    print(f"{user} → {count} times")
```
