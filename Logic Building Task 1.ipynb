```python
#1. User Login Check

username = input("Enter Username: ")
password = input("Enter Password: ")

if username == "admin" and password == "1234":
    print("Login Successful")
else :
    print("Invalid Credentials")
```

```python
#2. Pass / Fail Analyzer
marks = [45 , 78 , 90 , 30 , 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1
print("Total Students : ", len(marks))
print("Passed Students :", pass_count)
print("Failed Students :", fail_count)
```

```python
#3. Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = [name.strip().lower() for name in names]
# strip() => Removes leading and trailing spaces
# lower() => Converts text to lowercase

print(cleaned_names)
```

```python
# 4. Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    length = len(message)
    print(f"Message: {message}")
    print(f"Length: {length}")
    #The f stands for Formatted String Literal (also called an f-string).
    #It allows you to insert variables directly inside a string using {}.

    if length > 10:
        print("This message is longer than 10 characters.")

        print()
```

```python
#5. Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = logs.count("ERROR")
#Using count , inbuilt method in python.

print("Total Error entries: " ,error_count)
```
